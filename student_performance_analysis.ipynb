```python
# STUDENT PERFORMANCE ANALYSIS - COMPLETE CODE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better looking charts
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Load data
df = pd.read_csv('student_lifestyle_dataset.csv')

# Display basic info
print("="*60)
print("STUDENT LIFESTYLE DATASET ANALYSIS")
print("="*60)
print(f"Dataset shape: {df.shape[0]} students, {df.shape[1]} features")
print("\nFirst few rows:")
print(df.head())
print("\n" + "="*60)

# 1. SUMMARY STATISTICS
print("\n📊 SUMMARY STATISTICS")
print("-"*40)
print(df.describe())
print("\n" + "="*60)

# 2. CORRELATION ANALYSIS
print("\n🔗 CORRELATION BETWEEN VARIABLES")
print("-"*40)
# Select only numeric columns for correlation
numeric_cols = ['Study_Hours_Per_Day', 'Extracurricular_Hours_Per_Day', 
                'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day', 
                'Physical_Activity_Hours_Per_Day', 'GPA']
correlation_matrix = df[numeric_cols].corr()

# Display correlation with GPA (most important)
print("\nCorrelation with GPA:")
gpa_corr = correlation_matrix['GPA'].sort_values(ascending=False)
print(gpa_corr)
print("\n" + "="*60)

# 3. VISUALIZATIONS
print("\n📈 GENERATING VISUALIZATIONS...")

# Figure 1: Distribution of Study Hours
plt.figure(figsize=(10, 6))
sns.histplot(df['Study_Hours_Per_Day'], bins=20, kde=True, color='blue')
plt.title('Distribution of Daily Study Hours', fontsize=16)
plt.xlabel('Study Hours per Day')
plt.ylabel('Number of Students')
plt.axvline(df['Study_Hours_Per_Day'].mean(), color='red', linestyle='--', label=f'Mean: {df["Study_Hours_Per_Day"].mean():.2f} hrs')
plt.legend()
plt.tight_layout()
plt.show()

# Figure 2: Study Hours vs GPA
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Study_Hours_Per_Day', y='GPA', hue='Stress_Level', alpha=0.7, s=100)
plt.title('Study Hours vs GPA (colored by Stress Level)', fontsize=16)
plt.xlabel('Study Hours per Day')
plt.ylabel('GPA')
plt.legend(title='Stress Level')
plt.tight_layout()
plt.show()

# Figure 3: Sleep Hours vs GPA
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sleep_Hours_Per_Day', y='GPA', hue='Stress_Level', alpha=0.7, s=100)
plt.title('Sleep Hours vs GPA (colored by Stress Level)', fontsize=16)
plt.xlabel('Sleep Hours per Day')
plt.ylabel('GPA')
plt.legend(title='Stress Level')
plt.tight_layout()
plt.show()

# Figure 4: Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Between Lifestyle Factors and GPA', fontsize=16)
plt.tight_layout()
plt.show()

# Figure 5: Stress Level Distribution
plt.figure(figsize=(10, 6))
stress_counts = df['Stress_Level'].value_counts()
colors = ['#ff9999', '#66b3ff', '#99ff99']
plt.pie(stress_counts.values, labels=stress_counts.index, autopct='%1.1f%%', 
        colors=colors, startangle=90, explode=[0.05, 0.05, 0.05])
plt.title('Distribution of Stress Levels Among Students', fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()

# Figure 6: Average GPA by Stress Level
plt.figure(figsize=(10, 6))
avg_gpa_by_stress = df.groupby('Stress_Level')['GPA'].mean().sort_values()
avg_gpa_by_stress.plot(kind='barh', color=['#ff9999', '#ffcc99', '#99ff99'])
plt.title('Average GPA by Stress Level', fontsize=16)
plt.xlabel('Average GPA')
plt.ylabel('Stress Level')
for i, v in enumerate(avg_gpa_by_stress):
    plt.text(v + 0.01, i, f'{v:.3f}', va='center')
plt.tight_layout()
plt.show()

# Figure 7: Boxplot - GPA by Stress Level
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Stress_Level', y='GPA', palette=['#ff9999', '#ffcc99', '#99ff99'])
plt.title('GPA Distribution by Stress Level', fontsize=16)
plt.xlabel('Stress Level')
plt.ylabel('GPA')
plt.tight_layout()
plt.show()

# 4. KEY INSIGHTS
print("\n🔍 KEY INSIGHTS")
print("-"*40)

# Insight 1: Top factors affecting GPA
print("\n✅ Top factors correlated with GPA:")
for factor, corr in gpa_corr.items():
    if factor != 'GPA':
        strength = "strong" if abs(corr) > 0.5 else "moderate" if abs(corr) > 0.3 else "weak"
        direction = "positive" if corr > 0 else "negative"
        print(f"   • {factor}: {corr:.3f} ({strength} {direction} correlation)")

# Insight 2: Sleep recommendations
print("\n✅ Sleep analysis:")
avg_gpa_good_sleep = df[df['Sleep_Hours_Per_Day'] >= 7]['GPA'].mean()
avg_gpa_poor_sleep = df[df['Sleep_Hours_Per_Day'] < 6]['GPA'].mean()
print(f"   • Students sleeping 7+ hours: Avg GPA = {avg_gpa_good_sleep:.3f}")
print(f"   • Students sleeping <6 hours: Avg GPA = {avg_gpa_poor_sleep:.3f}")
print(f"   • Difference: {avg_gpa_good_sleep - avg_gpa_poor_sleep:.3f} points")

# Insight 3: Study time recommendations
print("\n✅ Study time analysis:")
high_gpa = df[df['GPA'] > 3.5]['Study_Hours_Per_Day'].mean()
low_gpa = df[df['GPA'] < 2.5]['Study_Hours_Per_Day'].mean()
print(f"   • High GPA students (>3.5): Study {high_gpa:.2f} hrs/day")
print(f"   • Low GPA students (<2.5): Study {low_gpa:.2f} hrs/day")

# Insight 4: Stress impact
print("\n✅ Stress level impact:")
stress_impact = df.groupby('Stress_Level')['GPA'].agg(['mean', 'count'])
for level in stress_impact.index:
    print(f"   • {level} stress: GPA = {stress_impact.loc[level, 'mean']:.3f} ({int(stress_impact.loc[level, 'count'])} students)")

print("\n" + "="*60)
print("✅ ANALYSIS COMPLETE!")
```
